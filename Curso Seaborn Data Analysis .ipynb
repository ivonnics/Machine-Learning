```python
import seaborn as sns
df = sns.load_dataset('iris')
#get a snippet of our data to look at data formats and size
df.head()
#make pairplots
sns.pairplot(df, vars = ["sepal_length", "sepal_width","petal_length", "petal_width" ], dropna = True)
```
