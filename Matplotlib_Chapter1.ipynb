```
import numpy
import matplotlib.pyplot as plot

def plot_slope(X, Y):
	Xs = X[1:] - X[:-1]
	Ys = Y[1:] - Y[:-1]
	plot.plot(X[1:], Ys / Xs)

X = numpy.linspace(-3, 3, 100)
Y = numpy.exp(-X ** 2)

plot.plot(X, Y)
plot_slope(X, Y)
plot.show()
```

```
import matplotlib.pyplot as plot

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

gap = .8 / len(data)
for i, row in enumerate(data):
	plot.bar([x + i * gap for x in range(len(row))], row, width = gap)

plot.show()

```

```
import matplotlib.pyplot as plot

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

color_list = ['b', 'g', 'r']

gap = .8 / len(data)
for i, row in enumerate(data):
	plot.bar([x + i * gap for x in range(len(row))], row, width = gap, color = color_list[i])

plot.show()
```

```
import numpy
import matplotlib.pyplot as plot

data = numpy.array([[5., 30., 45., 22.],
                    [5., 25., 50., 20.],
                    [1.,  2.,  1.,  1.]])

color_list = ['b', 'g', 'r']

X = numpy.arange(data.shape[1])
for i in range(data.shape[0]):
	S = numpy.sum(data[:i], axis = 0)
	plot.bar(X, data[i], bottom = S, color = color_list[i % len(color_list)])

plot.show()
```

```
import numpy
import matplotlib.pyplot as plot

data = numpy.array([[5., 30., 45., 22.],
                    [5., 25., 50., 20.],
                    [1.,  2.,  1.,  1.]])

color_list = ['b', 'g', 'r']
X = range(data.shape[1])
for i, col in enumerate(color_list):
	plot.bar(X, data[i], bottom = numpy.sum(data[:i], axis = 0), color = col)

plot.show()
```

```
import random
import matplotlib.pyplot as plot
import matplotlib.tri as tri


count = 100
X = [random.random() for i in range(count)]
Y = [random.random() for i in range(count)]

triangles = tri.Triangulation(X, Y)

plot.triplot(triangles, 'bo-')
plot.show()
```

```
import random
import matplotlib.pyplot as plot
import matplotlib.tri as tri


count = 100
X = [random.random() for i in range(count)]
Y = [random.random() for i in range(count)]
Z = [0.] * count

triangles = tri.Triangulation(X, Y)

plot.tripcolor(triangles, Z, edgecolors='w')
plot.show()
```
