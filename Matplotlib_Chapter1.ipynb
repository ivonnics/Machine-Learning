```
import numpy
import matplotlib.pyplot as plot

def plot_slope(X, Y):
	Xs = X[1:] - X[:-1]
	Ys = Y[1:] - Y[:-1]
	plot.plot(X[1:], Ys / Xs)

X = numpy.linspace(-3, 3, 100)
Y = numpy.exp(-X ** 2)

plot.plot(X, Y)
plot_slope(X, Y)
plot.show()
```

```
import matplotlib.pyplot as plot

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

gap = .8 / len(data)
for i, row in enumerate(data):
	plot.bar([x + i * gap for x in range(len(row))], row, width = gap)

plot.show()

```

```
import matplotlib.pyplot as plot

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

color_list = ['b', 'g', 'r']

gap = .8 / len(data)
for i, row in enumerate(data):
	plot.bar([x + i * gap for x in range(len(row))], row, width = gap, color = color_list[i])

plot.show()
```

```
import numpy
import matplotlib.pyplot as plot

data = numpy.array([[5., 30., 45., 22.],
                    [5., 25., 50., 20.],
                    [1.,  2.,  1.,  1.]])

color_list = ['b', 'g', 'r']

X = numpy.arange(data.shape[1])
for i in range(data.shape[0]):
	S = numpy.sum(data[:i], axis = 0)
	plot.bar(X, data[i], bottom = S, color = color_list[i % len(color_list)])

plot.show()
```

```
import numpy
import matplotlib.pyplot as plot

data = numpy.array([[5., 30., 45., 22.],
                    [5., 25., 50., 20.],
                    [1.,  2.,  1.,  1.]])

color_list = ['b', 'g', 'r']
X = range(data.shape[1])
for i, col in enumerate(color_list):
	plot.bar(X, data[i], bottom = numpy.sum(data[:i], axis = 0), color = col)

plot.show()
```

```
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')
fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], lw=3)
def init():
  line.set_data([], [])
  return line,
def animate(i):
 x = np.linspace(0, 4, 1000)
 y = np.sin(2 * np.pi * (x - 0.01 * i))
 line.set_data(x, y)
 return line,
anim = FuncAnimation(fig, animate, init_func=init,
frames=200, interval=20, blit=True)
#anim.save('sine_wave.gif', writer='imagemagick')
anim.save('sine_wave.gif', writer='pillow')
```
