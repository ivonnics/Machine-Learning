```
import pandas as pd

#Load data
url = "https://en.wikipedia.org/wiki/List_of_countries_by_number_of_military_and_paramilitary_personnel"
#url = "C:\\Users\\ivonnics\\Documents\\JOSE LUIS\\Cursos Cursera\\Machine Learning\\CJD2.csv"
dataset = pd.read_html(url)
Tabla=dataset[0]
Tabla.head(20)
```

```
df=Tabla
df = df.drop(0, 1)
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df['Country'] = df['Country'].str.replace("\[.*$", '')
df["Active military"]=df["Active military"].astype(int)
df["Reserve military"]=df["Reserve military"].astype(int)
df["Paramilitary"]=df["Paramilitary"].astype(int)
df["Total"]=df["Total"].astype(int)
df["Per 1000 capita(total)"]=df["Per 1000 capita(total)"].astype(float)
df["Per 1000 capita(active)"]=df["Per 1000 capita(active)"].astype(float)
print(df.shape)
print(df.dtypes)
df.head()
```

```
#Load data
url = "https://meta.wikimedia.org/wiki/List_of_countries_by_regional_classification"
#url = "C:\\Users\\ivonnics\\Documents\\JOSE LUIS\\Cursos Cursera\\Machine Learning\\CJD2.csv"
dataset = pd.read_html(url)
Tabla=dataset[0]
print(Tabla.shape)
print(Tabla.dtypes)
Tabla.head()
```

```
df2=Tabla
new_header = df2.iloc[0] #grab the first row for the header
df2 = df2[1:] #take the data less the header row
df2.columns = new_header #set the header row as the df header
df2 = df2.rename(columns={'Global South': 'Hemisphere'})
df2.head()
```

```
left = df
right = df2
df3=pd.merge(left, right, on='Country')
df4=df4.groupby('Region').sum()
df4
```
