## Matplotlib

```python
import matplotlib.pyplot as plt

plt.hist(df['petal_length'])

plt.title('Matplotlib Histogram')
plt.xlabel('petal length (cm)')
plt.savefig('Hist.png')
```

```python
import matplotlib.pyplot as plt

colors = {
            'setosa': 'r',
            'versicolor': 'g',
            'virginica': 'b'
         }
 
for specie in df['species'].unique():
    x = df[df['species'] == specie]['sepal_length']
    y = df[df['species'] == specie]['sepal_width']
    plt.scatter(x, y, c=colors[specie], edgecolors='none', s=30, label=specie)
 
plt.title('Matplotlib scatter plot')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
plt.savefig('Scatter.png')
```

## Seaborn

```python
import seaborn as sns

sns.pairplot(df, hue="species")

plt.savefig("Pairplot.png")
```

```python
import seaborn as sns
import pandas as pd

df = pd.melt(df, "species", var_name="measurement")

sns.stripplot(x="value", y="measurement", hue="species", data=df, alpha=.5)

plt.legend(loc="lower right")
plt.savefig("Stripplot.png")
```

```python
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
df = sns.load_dataset('iris')

data = {'x_values': [1, 2, 3, 4, 5],
        'y_values': [6, 7, 2, 3, 6]}

source = ColumnDataSource(data=data)

p = figure()
p.circle(x='x_values', y='y_values', source=source)
show(p)
```

## Plotly

```python
import plotly.plotly as py
import plotly.graph_objs as go

measurements = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

colors = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c"]
       
fig = {
    "data": sum(
        [
            [
                {
                    "type": 'violin',
                    "x": df[df['species'] == specie]['species'],
                    "y": df[df['species'] == specie][measurement],
                    "name": measurement,
                    "legendgroup": measurement,
                    "showlegend": specie == 'setosa',
                    "visible": ("legendonly" if i != 0 else True),
                    "box": {
                        "visible": True
                    },
                    "meanline": {
                        "visible": True
                    },
                    "line": {
                        "color": colors[i]
                    }
                }
            for i, measurement in enumerate(measurements)] 
        for specie in df['species'].unique()], 
        []
    ),
    "layout" : {
        "title": "Iris Dataset Distribution",
    }
}

py.iplot(fig)
```

### Observa que tanto los gráficos de Bokeh como los de Plotly son interactivos y abren un webserver, local en el caso de Plotly, y externo en el caso de Bokeh, donde puedes interactuar mejor con ellos...
