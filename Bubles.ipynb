```
#  Taken from: https://medium.com/@rinu.gour123/python-charts-bubble-3d-charts-with-properties-of-chart-f24a99f72469
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
x=np.random.rand(30)
y=np.random.rand(30)
z=np.random.rand(50)
plt.scatter(x,y,s=z*777) #s is for global size
plt.show()
```

```
plt.scatter(x,y,s=z*777,c='Chartreuse')
plt.show()
```

```
colors=np.random.rand(30)
plt.scatter(x,y,s=z*777,c=colors,alpha=0.5) #alpha for transparency
plt.show()
```

```
plt.scatter(x,y,s=z*300,marker='D')
plt.show()
```

```
plt.scatter(x,y,s=z*300,marker='<',color='brown')
plt.show()
```

```
plt.scatter(x,y,s=z*500,marker='*',color='pink')
plt.show()
```

```
plt.scatter(x,y,s=z*500,marker='8',color='red')
plt.show()
```

```
plt.scatter(x,y,s=z*4000,c='beige',alpha=0.4,linewidth=7)
plt.show()
```

```
plt.scatter(x,y,s=z*800,c='beige',linewidth=7,edgecolors='brown')
plt.show()
```

```
plt.scatter(x,y,s=z*2000,c=x,cmap='Greys',alpha=0.4,edgecolors='grey',linewidth=2)
plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D
df=pd.DataFrame({'X':range(1,111),'Y':np.random.randn(110)*13+range(1,111), 'Z':(np.random.randn(110)*13+range(1,111))*2})
fig=plt.figure()
ax=fig.add_subplot(121,projection='3d')
ax.scatter(df['X'],df['Y'],df['Z'],c='pink',s=60)
ax.view_init(27,200)
plt.show()
```

```
x=np.arange(0,7)
from scipy import sin
y=sin(x)
plt.plot(x,y)
plt.annotate(xy=[3.5,0],s='The curve') #You can add the annotations before plotting if you want
plt.show()
```

```
 x=np.arange(0,7)
y=sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
plt.legend(['Sine','Cosine'],loc=3)
plt.annotate(xy=[3.5,0],s='The curves') #You can add the annotations before plotting if you want
plt.show()
```
